```python
import pandas as pd
import seaborn as sns

# =========================
# MUAT DATASET PENGUINS
# =========================

df = sns.load_dataset('penguins')

# Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())

# =========================
# FILTERING
# Penguin Betina dari Spesies Adelie
# =========================

adelie_female = df[
    (df['species'] == 'Adelie') &
    (df['sex'] == 'Female')
]

print(
    adelie_female['body_mass_g'].mean(),
    adelie_female['bill_length_mm'].mean()
)

# =========================
# GROUPBY
# Rata-rata Massa Tubuh per Spesies
# =========================

hasil = df.groupby('species').agg(
    total=('species', 'count'),
    rata_massa=('body_mass_g', 'mean')
).round(2)

print(hasil)

# =========================
# JAWABAN KUIS
# =========================

# Q1
q1_question = "Q1 Berapa jumlah baris dan kolom dataset Penguins?"
q1_answer = f"{df.shape[0]} baris, {df.shape[1]} kolom"

# Q2
q2 = (df.isnull().sum() / len(df) * 100).round(2)

# Q3
q3_question = "Q3 Berapa rata-rata massa tubuh seluruh penguin?"
q3_answer = f"{df['body_mass_g'].mean():.2f} gram"

# Q4
q4_question = "Q4 Berapa rata-rata panjang paruh penguin betina spesies Adelie?"
q4_answer = f"{adelie_female['bill_length_mm'].mean():.2f} mm"

# Q5
q5_question = "Q5 Spesies mana yang memiliki rata-rata massa tubuh tertinggi?"
q5_answer = df.groupby('species')['body_mass_g'].mean().idxmax()

# Q6
q6 = df.groupby('species')['body_mass_g'].mean().round(2)

# =========================
# TAMPILKAN DALAM TABEL
# =========================

hasil_tabel = pd.DataFrame({
    "Pertanyaan": [
        q1_question,
        q3_question,
        q4_question,
        q5_question
    ],
    "Jawaban": [
        q1_answer,
        q3_answer,
        q4_answer,
        q5_answer
    ]
})

print("\n===== JAWABAN Q1, Q3, Q4, Q5 =====")
display(hasil_tabel)

print("\n===== PERSENTASE MISSING VALUE (Q2) =====")
display(q2.to_frame(name="Persentase Missing (%)"))

print("\n===== RATA-RATA MASSA TUBUH PER SPESIES (Q6) =====")
display(q6.to_frame(name="Rata-rata Massa (gram)"))
```

Kesimpulan: Praktikum ini mempelajari cara menggunakan Pandas untuk memuat dataset, melakukan eksplorasi data awal, menghitung persentase data yang hilang (missing values), melakukan filtering berdasarkan kondisi tertentu, serta menggunakan fungsi groupby() untuk memperoleh informasi yang lebih mendalam dari setiap kelompok data. Melalui analisis ini dapat diketahui karakteristik dataset, pola pada data, serta perbedaan antar kelompok yang dianalisis. Praktikum ini membantu memahami teknik dasar pengolahan dan analisis data yang menjadi fondasi dalam Data Science.
