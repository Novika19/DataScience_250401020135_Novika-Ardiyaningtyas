```python
# SEL 1: Output sederhana
print("=== BELAJAR DASAR PYTHON ===")

# SEL 2: Variabel
nama = "Novika Ardiyaningtyas"
nim = "250401020135"
kelas = "IF401"

print("Nama :", nama)
print("NIM  :", nim)
print("Kelas:", kelas)

# SEL 3: List Mata Kuliah
mata_kuliah = [
    "Data Science Dasar",
    "Algoritma dan Pemrograman",
    "Basis Data",
    "Matematika Diskrit",
    "Jaringan Komputer"
]

# SEL 4: List Nilai
nilai = [90, 85, 88, 92, 87]

print("\nDaftar Mata Kuliah dan Nilai:")
for i in range(len(mata_kuliah)):
    print(f"{i+1}. {mata_kuliah[i]} = {nilai[i]}")


# SEL 65 Fungsi Sederhana
def tampilkan_identitas(nama, nim, kelas):
    return f"{nama} | {nim} | {kelas}"

# SEL 6: Memanggil Fungsi
print("\nIdentitas Mahasiswa:")
print(tampilkan_identitas(nama, nim, kelas))
```

Kesimpulan: Praktik ini membahas dasar-dasar pemrograman Python, termasuk penggunaan fungsi "print()" untuk menampilkan output, variabel untuk menyimpan data, list untuk menyimpan kumpulan informasi, dan perulangan "for" untuk menampilkan data secara berurutan. Data mata kuliah dan nilai disimpan dalam list, kemudian ditampilkan menggunakan perulangan sehingga lebih terstruktur dan mudah dibaca.
